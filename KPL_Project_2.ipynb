- 보행환경개선사업을 통한 보행의 회복은 단순히 보행자의 안전 뿐만 아니라 도시 공간의 활력을 불러올 수 있음.

- 더 나아가 보행활동을 통한 집합적인 장소경험의 공유는 집합적인 지역의 이미지를 공유하여 지역공동체를 묶어주는 ‘공동체의식’을 이끌어 냄. 또한 기후변화, 저탄소 정책, 보행약자의 이동권 보장 등 다양한 사회적 요구를 수렴하여 지속 가능한 도시공간체계로의 변화를 도모하기 위해서는 걷고 싶은 도시가 가지는 잠재력을 고려해야 함.

  - _원활한 보행 이동이 보장된다면 보행과 대중교통만으로도 구 내 이동이 가능하기 때문에 불법주정차 및 상습적인 정체 현상은 감소할 것으로 기대됨. 그리고 이면도로가 많은 광진구에서 보행환경개선사업이 원활히 진행된다면 휠체어 등을 이용하는 장애인 등 노약자의 보도 이동권이 보장될 것으로 사료됨._

- 광진구가 지속가능한 도시로서 발돋움하기 위해서는 보행환경개선사업이 일회성 사업으로 전락하지 않도록 지속적인 관심과 투자, 철저한 사후 관리가 필요함.

- 본 프로젝트는 기존 광진구에서 진행한 두 곳과 다른 자치구의 보행환경개선지구를 바탕으로 사업 전과 후의 보행환경이 정량적으로 얼만큼 개선되었는지 파악하고자 함. 이를 위해 **보행환경을 평가할 수 있는 변수를 설정**하고 통계기법을 활용하고자 함. 이후 광진구 내 보행환경 개선이 시급한 구역을 선정하고 세부적인 효과까지 도출하는 프로젝트를 진행하고자 함.

      2. 기존 보행환경 개선사업 지역의 폴리곤과 블록이 중첩되는 부분을 추출함 (QGIS 내 '교차영역' 기능 사용)
      3. 중첩 구역의 dbf 파일을 csv 파일로 불러와 추정매출액 계산함
    




### **사용한 코드 (명동 지역 예시)**
#### **'13-'15 데이터와 '16-'19 모두 동일한 코드를 사용하였음**


```
import pandas as pd
from dbfread import DBF

# 추정매출액 txt 파일을 데이터프레임으로 불러옵니다.
df = pd.read_csv('th_ic_ecm_blck_selng_qu_2013_2015.txt', delimiter = '|', quotechar = " ' ", encoding = 'utf-8')

#sales_13_15는 df 중 필요한 컬럼만 추출하여 생성한 데이터프레임

# DBF 파일을 pandas 데이터프레임으로 읽어옵니다.
dbf = DBF('이태원 매출 블록.dbf', encoding='utf-8')

# 지역의 블록코드열을 읽어옵니다.
block_code = [record["BLCK_CD"] for record in dbf]

# comma로 구분된 문자열로 변환
block_code_str = ",".join(str(number)) for number in block_code)

# comma로 구분된 문자열을 정수형 리스트로 변환
blk_cd_list = [int(number) for number in block_code_str.split(",")]

# sales_13_15.csv 파일을 pandas 데이터프레임으로 읽어옵니다.
df_sales = pd.read_csv('sales_13_15.csv', encoding='cp949')

# 블록코드가 선택한 blk_cd_list에 속하는 데이터를 필터링합니다.
df_sales_selected = df_sales[df_sales['블록코드'].isin(blk_cd_list)]

# 필터링된 데이터에서 분기매출액을 총 합산합니다.
total_sales_1315 = df_sales_selected['분기매출액'].sum()

# 16~19년 매출도 위와 같은 코드 사용하여 진행

print (total_sales_1315)
print (total_sales_1619)
print (100*(total_sales_1619 - total_sales_1315)/total_sales_1315)
```
