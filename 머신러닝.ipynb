```python
# 머신러닝
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC, LinearSVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import Perceptron
from sklearn.linear_model import SGDClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
df=pd.read_excel('test1.xlsx')
df
```

```python
#데이터 스케일링

from sklearn.preprocessing import StandardScaler
df_new=df.iloc[:,1:]
# Standard scaler
df_scale = pd.DataFrame(StandardScaler().fit_transform(df_new), columns=df_new.columns, index = df_new.index)
df_scale
```

```python
# KMO Test (Kaiser-Meyer-Olkin Test)
## 변수들간 상관관계가 다른 변수에 의해 잘 설명되는 정도를 나타내는 값
## 이 값이 작으면 요인분석을 위한 변수들의 선정이 적절하지 못하다는 것을 의미
## 일반적으로 0.9 이상이면 아주 좋음, 0.8 이상이면 꽤 좋음, 0.7 이상이면 적당함, 0.6 이상이면 평범함, 0.5 미만이면 안 좋음
 
from factor_analyzer.factor_analyzer import calculate_kmo
kmo_all, kmo_model =calculate_kmo(df_scale)
kmo_model # 0.6이므로 사용가능
```

```python
df2=pd.read_excel('test2.xlsx')
df_test=df2.iloc[:,1:]
# Standard scaler
df_scale2 = pd.DataFrame(StandardScaler().fit_transform(df_test), columns=df_test.columns, index = df_test.index)
df_scale2
```

```python
#독립변수
X_train = df_scale.drop("양호여부", axis=1)
#종속변수
Y_train = df_scale["양호여부"]
#예측 대상 데이터 셋
X_test = df_scale2
```

```python
# Logistic Regression
logreg = LogisticRegression()
logreg.fit(X_train, Y_train)
Y_pred = logreg.predict(X_test)
res_log=Y_pred
acc_log = round(logreg.score(X_train, Y_train) * 100, 2)
acc_log
```

```python
# Support Vector Machines
svc = SVC()
svc.fit(X_train, Y_train)
Y_pred = svc.predict(X_test)
res_svc=Y_pred
acc_svc = round(svc.score(X_train, Y_train) * 100, 2)
acc_svc
```

```python
knn = KNeighborsClassifier(n_neighbors = 3)
knn.fit(X_train, Y_train)
Y_pred = knn.predict(X_test)
res_knn=Y_pred
acc_knn = round(knn.score(X_train, Y_train) * 100, 2)
acc_knn
```

```python
# Gaussian Naive Bayes

gaussian = GaussianNB()
gaussian.fit(X_train, Y_train)
Y_pred = gaussian.predict(X_test)
res_gaussian=Y_pred
acc_gaussian = round(gaussian.score(X_train, Y_train) * 100, 2)
acc_gaussian
```

```python
# Perceptron

perceptron = Perceptron()
perceptron.fit(X_train, Y_train)
Y_pred = perceptron.predict(X_test)
res_perceptron=Y_pred
acc_perceptron = round(perceptron.score(X_train, Y_train) * 100, 2)
acc_perceptron
```

```python
# Linear SVC

linear_svc = LinearSVC()
linear_svc.fit(X_train, Y_train)
Y_pred = linear_svc.predict(X_test)
res_linear_svc = Y_pred
acc_linear_svc = round(linear_svc.score(X_train, Y_train) * 100, 2)
acc_linear_svc
```

```python
# Decision Tree

decision_tree = DecisionTreeClassifier()
decision_tree.fit(X_train, Y_train)
Y_pred = decision_tree.predict(X_test)
res_decision_tree=Y_pred
acc_decision_tree = round(decision_tree.score(X_train, Y_train) * 100, 2)
acc_decision_tree
```

```python
# Random Forest

random_forest = RandomForestClassifier(n_estimators=100)
random_forest.fit(X_train, Y_train)
Y_pred = random_forest.predict(X_test)
res_random_forest=Y_pred
random_forest.score(X_train, Y_train)
acc_random_forest = round(random_forest.score(X_train, Y_train) * 100, 2)
acc_random_forest
```

```python
# Stochastic Gradient Descent

sgd = SGDClassifier()
sgd.fit(X_train, Y_train)
Y_pred = sgd.predict(X_test)
res_sgd=Y_pred
acc_sgd = round(sgd.score(X_train, Y_train) * 100, 2)
acc_sgd
```
